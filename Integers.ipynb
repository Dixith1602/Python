```python
print("Question 1: Calculate the product of first 10 natural numbers")

product = 1
for i in range (1,11):
    product *= i

print(f"The product of first 10 numbers is: {product}")
```

```python
print("Question 2: Find the remainder when 156 is divided by 7")
print(f"The remainder is: {156%7}")
```
